```
def zadanie1a(text):
  a = 'państwo'
  b = 'stolica'
  case = text.lower()
  change = list(case.replace(a, b))
  
  indicator = []
  for i, l in enumerate(text):
    if l.isupper():
      indicator.append(i)

  for i in indicator:
    change[i] = change[i].upper()

  return ''.join(change)

zadanie1a("Państwo pAństwo PAńSTWO")
```

```
def zadanie1b(text):
  a = 'kraj'
  b = 'państwo'
  case = text.lower()
  change = case.replace(a, b)

  spl_text = text.split()
  spl_change = change.split()

  for i in spl_text:
    countOneText = len(i)
  for e in spl_change:
    countOneChange = len(e)

  for z in range(len(spl_text)):
    spl_text[z] = spl_text[z] + 'x' * ( countOneChange - countOneText)    

  text = ' '.join(spl_text)
  change = list(' '.join(spl_change))

  indicator = []
  for i, l in enumerate(text):
    if l.isupper():
      indicator.append(i)

  for i in indicator:
    change[i] = change[i].upper()
  
  return ''.join(change)

zadanie1b("Kraj kraj krAJ")
```

```
def zadanie2a():
  table = [
    { "date": "2019-11-02", "rate": 4.3422 },
    { "date": "2019-11-03", "rate": 4.2210 },
    { "date": "2019-11-04", "rate": 4.3455 },
    { "date": "2019-11-05", "rate": 4.3456 },
    { "date": "2019-11-06", "rate": 4.2311 }
  ]
  min_values = (min(x['rate'] for x in table))
  max_values = (max(x['rate'] for x in table))

  index_minimum = [i for i, n in enumerate(table) if n['rate'] == min_values][0]
  index_maximum = [i for i, n in enumerate(table) if n['rate'] == max_values][0]
  
  for i, k in enumerate(table):
    if i==index_minimum:
      resultMin = k
    if i==index_maximum:
      resultMax = k
  result = { "min" : resultMin, "max" : resultMax }
  return result

zadanie2a()
```

```
def zadanie2b():
  table = [
    { "date": "2019-11-02", "rate": 4.3422 },
    { "date": "2019-11-03", "rate": 4.2210 },
    { "date": "2019-11-04", "rate": 4.3455 },
    { "date": "2019-11-05", "rate": 4.3456 },
    { "date": "2019-11-06", "rate": 4.2311 }
  ]
  max_l = max(table, key=lambda x: x['rate'])
  min_l = min(table, key=lambda x: x['rate'])

  result = {"max" : max_l, "min": min_l }
  return result

zadanie2b()
```
